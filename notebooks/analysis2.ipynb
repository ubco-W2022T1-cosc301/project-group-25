```python
heatmapDf = df[["Goals", "Assists", "Blocks", "BlkSh", "BlkShSv", "BlkPass", "SCA", "SoT", "Tkl", "AerWon"]]
heatmapDf= heatmapDf.rename(columns={"BlkSh":"Blocked Shot","BlkPass":"Blocked Pass", "BlkShSv":"Blocked Shot On Target","SoT":"Shots On Target","Tkl":"Tackles","AerWon":"Aerials Won", "SCA":"Shot Creating Actions"})
correlation = heatmapDf.corr()

fig, ax = plt.subplots(figsize=(8,8))  
ax = sns.heatmap(correlation,
            xticklabels = correlation.columns,
            yticklabels = correlation.columns,
            annot = True).set_title('Correlations Between Notable Offensive and Defensive Player Statistics')
plt.xlabel("Per 90 minutes")
plt.ylabel("Per 90 minutes")
```

## Task 4 Analysis

```python
df = load_and_process("../data/raw/Stats.csv")
dfl = df.groupby(["League"]).mean().reset_index()
dfl["InitiativePoints"] = dfl["SoT"] + dfl["SCA"] + dfl["Tkl"] + dfl["TklAtt3rd"]
dfl["DefensivePoints"] = dfl["Blocks"] + dfl["BlkSh"] + dfl["BlkShSv"]+ dfl["BlkPass"]
dfl["RecklessPoints"] = dfl["CrdY"] + dfl["RedCards"] + dfl["2CrdY"] + dfl["Fls"] + dfl["Fld"]
```

In order to compare the 5 different leagues, I have grouped player statistics that I believe are most relative when considering certain playstyles and player strengths on average.

Aggressive Initiative points:
* SoT : Shots on target (Does not include penalty kicks)
* SCA : Shot-creating actions
* Tkl : Number of players tackled
* TklAtt3rd : Tackles in attacking 1/3

A league with a higher number of shots on target, shot creating actions, tackles and tackles in attacking zones are characteristics of a league with a more aggressive playstyle and has players which take more initiative in their play.

```python
sns.barplot(y = dfl["League"], x = dfl["InitiativePoints"]).set_title('Average Initiative Points Received by Players in Each League Per 90 minutes')
plt.ylabel("League")
plt.xlabel("Average Initiative Points Aggregation")
```

#### From this graph we notice that Bundesliga and Ligue 1 have similar playstyles when it comes to aggressiveness, since these two leagues scored highest and have players who take the most offensive initiative on average.

Defensive points:

* Blocks : Number of times blocking the ball by standing in its path
* BlkSh : Number of times blocking a shot by standing in its path
* BlkShSv : Number of times blocking a shot that was on target, by standing in its path
* BlkPass : Number of times blocking a pass by standing in its path

A higher number of blocks, blocked shots, blocked shots on target, and blocked passes are characteristics of a league with better defensive play.

```python
sns.barplot(y = dfl["League"], x = dfl["DefensivePoints"]).set_title('Defensive Points Received by Players in Each League Per 90 minutes')
plt.ylabel("League")
plt.xlabel("Average Defensive Points Aggregation")
```

#### From this graph we can see that Serie A has the best defensive players on average, whereas Ligue 1 scored the lowest meaning their leagues playstyle constitutes less defensive play.

Reckless points:
* CrdY : Yellow cards
* Red Cards
* 2CrdY : Second yellow card
* Fls : Fouls committed
* Fld : Fouls drawn

A higher number of yellow cards, second yellow cards, red cards, fouls committed, and fouls drawn are characteristics of a league with a more reckless playstyle and less discipline.


```python
sns.barplot(y = dfl["League"], x = dfl["RecklessPoints"]).set_title('Reckless Points Received by Players in Each League Per 90 minutes')
plt.ylabel("League")
plt.xlabel("Average Reckless Points Aggregation")
```

#### From this observation we can see that La Liga scored the highest on this point system, concluding that this league was the least disciplined and has the most reckless playstyle, whereas the Premier League has the most disciplined players on average. 

### Observations

* La Liga's football players scored the highest for reckless points on average while also scoring second lowest for defensive points. Bundesliga's football players scored the highest on average for defensive points, while also scoring the highest for initiative points. La Liga has the oldest football players on average, and these players also scored the least number of goals per 90 minutes on average.

* Bundesliga and Ligue 1 have similar playstyles when it comes to aggressiveness, since these two leagues scored highest on the initiative points aggregation, and thus have players who take the most offensive initiative on average. Bundesliga has the best defensive players on average, whereas Ligue 1 scored the lowest meaning their leagues playstyle constitutes less defensive play. 

* La Liga scored the highest on the reckless points aggregation, concluding that this league was the least disciplined and has the most reckless playstyle, whereas the Premier League has the most disciplined players on average. Bundesliga players scored the most goals on average during the 2021-2022 Season. Ligue 1 players received the most red cards on average during the 2021-2022 Season


### Conclusion: Overall, players from the Bundesliga league have the best performance on average. Since players from this league on average per 90 minutes scored the most goals, have the most assists, scored the highest on the Press statistic, receive the least number of red cards, were awarded the most defensive points, initiative points, and scored among the lowest on the reckless points aggregation.
