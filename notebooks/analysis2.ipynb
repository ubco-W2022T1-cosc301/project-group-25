## Task 4 Analysis

```python
df = load_and_process("../data/raw/Stats.csv")
dfl = df.groupby(["League"]).mean().reset_index()
dfl["InitiativePoints"] = dfl["SoT"] + dfl["SCA"] + dfl["Tkl"] + dfl["TklAtt3rd"]
dfl["DefensivePoints"] = dfl["Blocks"] + dfl["BlkSh"] + dfl["BlkShSv"]+ dfl["BlkPass"]
dfl["RecklessPoints"] = dfl["CrdY"] + dfl["RedCards"] + dfl["2CrdY"] + dfl["Fls"] + dfl["Fld"]
```

In order to compare the 5 different league, I have grouped player statistics that I believe are most relative when considering certain playstyles and player strengths on average

Aggressive Initiative points:
* SoT : Shots on target (Does not include penalty kicks)
* SCA : Shot-creating actions
* Tkl : Number of players tackled
* TklAtt3rd : Tackles in attacking 1/3

A league with a higher number of shots on target, shot creating actions, tackles and tackles in atttacking zones are characteristics of a league with a more aggressive playstyle and has players which take more initiative in their play

```python
sns.barplot(y = dfl["League"], x = dfl["InitiativePoints"])
```

#### From this graph we notice that Bundesliga and Ligue 1 have similar playstyles when it comes to aggressiveness, since these two leagues scored highest and have players who take the most offensive initiave on average.

Defensive points:

* Blocks : Number of times blocking the ball by standing in its path
* BlkSh : Number of times blocking a shot by standing in its path
* BlkShSv : Number of times blocking a shot that was on target, by standing in its path
* BlkPass : Number of times blocking a pass by standing in its path

A higher number blocks, blocked shots, blocked shots on targed, and blocked passes are characteristics of a league with better defensive play. 

```python
sns.barplot(y = dfl["League"], x = dfl["DefensivePoints"])
```

#### From this graph we can see that Serie A has the best defensive players on average, whereas Ligue 1 scored the lowest meaning their leagues playtyle constitutes less defensive play. 

Reckless points:
* CrdY : Yellow cards
* RedCards
* 2CrdY : Second yellow card
* Fls : Fouls committed
* Fld : Fouls drawn

A higher number of yellow cards, second yellow cards, red cards, fouls committed, and fouls drawn are characteristics of a league with a more reckless playstyle and less discipline.


```python
sns.barplot(y = dfl["League"], x = dfl["RecklessPoints"])
```

#### From this observation we can see that La Liga scored the highest on this point system, concluding that this league was the least disciplined and has the most reckless playstyle, whereas Ligue 1 has the most disciplined players on average.
