## Task 2: Analysis Pipeline

1. Load data has already been completed above

2. Clean Data: I've decided to drop the following columns since I have deemed they are not relevant to my analysis:
* Carries : Number of times the player controlled the ball with their feet
* CarTotDist : Total distance, in yards, a player moved the ball while controlling it with their feet, in any direction
* CarPrgDist : Total distance, in yards, a player moved the ball while controlling it with their feet towards the opponent's goal
* CarProg : Carries that move the ball towards the opponent's goal at least 5 yards, or any carry into the penalty area
* Car3rd : Carries that enter the 1/3 of the pitch closest to the goal
* CPA : Carries into the 18-yard box
* OG : Own goals
I'll also rename the "Comp" column to League, and the "CrdR" column to RedCards since that name makes more sense to me intuitively

```python
df_raw = pd.read_csv("../data/raw/Stats.csv",encoding="ISO-8859-1", delimiter = ";")
df_clean = df.copy().drop(["Carries","CarTotDist", "CarPrgDist", "CarProg", "Car3rd", "CPA", "OG"], axis=1).dropna(axis=0)
df_clean.reset_index()
df_clean.rename(columns={"Comp":"League","CrdR":"RedCards"}, inplace=True)
df_clean.head()
```
